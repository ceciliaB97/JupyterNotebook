```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv("data_cleaning.csv")
df.head(5)
```

```python
to_drop = ['cp',
           'fbs',
           'restecg',
           'thalach',
           'exang',
           'oldpeak',
           'slope',
           'thal',
           'target',
           'ca']
df.drop(to_drop, inplace = True, axis = 1)
df.head(5)
```

```python
news_name = {'age':'Age',
             'sex':'Sex',
             'trestbps':'Bps',
             'chol':'Cholesterol'
            }
df.rename(columns = news_name, inplace = True)
df.head()
```

```python
replace_values = {0:'F', 1:'M'}
df = df.replace({'Sex': replace_values})
df.head()
```
